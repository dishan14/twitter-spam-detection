```python
from nltk.corpus import stopwords
from nltk.tokenize import TweetTokenizer 
from nltk.stem import WordNetLemmatizer
def clean(word):
    ttk=TweetTokenizer()
    nlp = WordNetLemmatizer()
    li=[]
    tokenised_word=ttk.tokenize(word)
    filtered_words = [word for word in tokenised_word if word not in stopwords.words('english')]
    stem_words=[nlp.lemmatize(i) for i in filtered_words]
    return(' '.join(e.lower()for e in stem_words))
    
```

```python
import re
def remove(sentence):
    li=re.sub(r'https://[A-Za-z0-9./]+','',sentence)
    li1=re.sub(r'http://[A-Za-z0-9./]+','',li)
    r=(re.sub(r'@[A-Za-z0-9]+','',li1))
    r1=re.sub(r'[^A-Za-z0-9]+',' ',r)
    return(r1)
```
